```python
budget, us_gross, world_gross = np.genfromtxt("MovieData.csv", delimiter='\t', 
        skip_header=1, usecols=(3,4,5), unpack=True, missing="Unknown")
```

```python
print "Mean movie budget:", budget.mean()
print "Movie budget standard deviation:", budget.std()
print "Median movie budget:", np.median(budget) # Note, no built-in mean function
```

```python
profit = (us_gross + world_gross)/budget
print "Mean profit: ", profit.mean()
```

```python
print profit
print np.isnan(profit)
```

```python
budget_clean = budget[~np.isnan(profit)]
profit_clean = profit[~np.isnan(profit)]
print profit_clean.max()
print profit_clean.mean()
print profit_clean.min()
```

```python
plt.scatter(budget_clean, profit_clean)
plt.ylim(0,20)
plt.xlim(0,3e8)
plt.xlabel("Budget")
plt.ylabel("Profit")
```

```python
linear_model = st.linregress(budget_clean,profit_clean)
print linear_model
```

```python
model_prediction = linear_model[1] + linear_model[0] * budget_clean
```

```python
fig, subplot = plt.subplots()
subplot.scatter(budget_clean, profit_clean)
subplot.plot(budget_clean, model_prediction, 'r')

subplot.set_xlabel("Budget")
subplot.set_ylabel("Profit")

subplot.set_ylim(0,20)
subplot.set_xlim(0,3e8)
subplot.legend(['Data', 'Linear Model'], loc=(1.03,0))
```

```python
profit_bins = np.histogram(profit_clean, bins=np.linspace(0,20,21))
print len(profit_bins[0])
```

```python
fig1 = plt.figure()
subplot = plt.subplot(111)

graph = subplot.scatter(np.linspace(0,20,20),profit_bins[0])
subplot.set_xlim(0,20)
subplot.set_ylim(0,400)
```

```python
# Slope, intercept, r-val, p-val, stderr
linear_model = st.linregress(np.linspace(0,20,20),profit_bins[0])
print linear_model
```

```python
model_predict = linear_model[1] + linear_model[0]*np.linspace(0,20,20)
fig, subplot = plt.subplots()
subplot.scatter(np.linspace(0,20,20), profit_bins[0])
plot = subplot.plot(np.linspace(0,20,20), model_predict)
subplot.set_xlim(0,20)
subplot.set_ylim(0,400)
subplot.legend(['Data', 'Linear Model'], loc=(1.03,0))
```

```python
fig1 = plt.figure()
subplot = plt.subplot(111)

graph = subplot.scatter(np.linspace(0,20,20),profit_bins[0])
subplot.set_xscale('log')
subplot.set_yscale('log')
```

```python
fig1 = plt.figure()
subplot = plt.subplot(111)
x_axis = np.linspace(0,20,20)+0.5
graph = subplot.loglog(np.linspace(0,20,20)+0.5,profit_bins[0], '.')
```

```python
empirical_pdf = 1.0*profit_bins[0]/np.sum(profit_bins[0])
empirical_cdf = np.cumsum(empirical_pdf)
plt.plot(empirical_cdf)
```

```python
plt.plot(empirical_cdf)

lognorm = st.lognorm
lognorm_fit = lognorm.fit(profit_bins[0])
lognorm_pdf = lognorm.pdf(x_axis, lognorm_fit[0])
lognorm_pdf = lognorm_pdf/np.sum(lognorm_pdf)
lognorm_cdf = np.cumsum(lognorm_pdf)

plt.plot(lognorm_cdf)
```

```python
pareto = st.pareto
pareto_fit = pareto.fit(profit_bins[0])
pareto_pdf = pareto.pdf(np.linspace(0,20,20)+0.5, pareto_fit[0])
pareto_pdf = pareto_pdf/np.sum(pareto_pdf)

pareto_cdf = np.cumsum(pareto_pdf)
plt.plot(pareto_cdf)
```

```python
fig = plt.figure(figsize=(8,8))
subplot = plt.subplot(111)

subplot.plot(empirical_cdf)
subplot.plot(lognorm_cdf)
subplot.plot(pareto_cdf)

subplot.legend(["Empirical", "Log Normal", "Pareto"])
```
